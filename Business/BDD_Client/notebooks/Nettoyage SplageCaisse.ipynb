```python
# SPLASH CAISSE

import pandas as pd
import os

# Chemins d'accès
PATH_ALFORNO = os.getenv("SPLASH_CAISSE_ALFORNO_PATH", r"C:\path\to\clients_alforno.csv")
PATH_BKF = os.getenv("SPLASH_CAISSE_BKF_PATH", r"C:\path\to\clients_BKF_pontault.csv")
OUTPUT_FILE = "clients_caisse_filtres.csv"

# Lecture des fichiers
df_alforno = pd.read_csv(PATH_ALFORNO)
df_bkf = pd.read_csv(PATH_BKF)

print("Alforno :", df_alforno.shape)
print("BKF :", df_bkf.shape)

# Ajout des colonnes source
df_alforno["Restaurant"] = "Alforno Pontault Combault"
df_bkf["Restaurant"] = "Bangkok Factory Pontault Combault"

df_alforno["Plateforme"] = "splash_caisse"
df_bkf["Plateforme"] = "splash_caisse"

# Harmonisation des noms de colonnes
colonnes_renommer = {
    "telephone": "Téléphone",
    "prenom": "Prénom",
    "creation": "Création",
    "derniere_visite": "Dernière Visite",
    "nb_visites": "Nombre de visites",
    "points": "Nombre de points"
}

df_alforno.rename(columns=colonnes_renommer, inplace=True)
df_bkf.rename(columns=colonnes_renommer, inplace=True)

# Fusion des deux bases
df_caisse = pd.concat([df_alforno, df_bkf], ignore_index=True)

# Fusion Prénom + nom
df_caisse["Nom"] = df_caisse["Prénom"].fillna("") + " " + df_caisse["nom"].fillna("")
df_caisse["Nom"] = df_caisse["Nom"].str.strip()

# Emails en minuscules
df_caisse["email"] = df_caisse["email"].str.lower()

# Suppression des lignes sans téléphone
df_caisse = df_caisse[df_caisse["Téléphone"].notna() & (df_caisse["Téléphone"] != "")]

# Colonnes finales
colonnes_finales = ["Nom", "email", "Téléphone", "Restaurant", "Plateforme"]
df_final = df_caisse[colonnes_finales]

# Résultat
print("Données nettoyées et filtrées :")
display(df_final.head())

# Sauvegarde
df_final.to_csv(OUTPUT_FILE, index=False, encoding="utf-8")
```
